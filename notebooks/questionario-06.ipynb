```python
from collections import Counter
from collections.abc import Sequence
from pathlib import Path
```

```python
import networkx as nx
import numpy as np
import pandas as pd
from networkx import Graph
from networkx.algorithms.community import (
    greedy_modularity_communities,
    louvain_communities,
    modularity,
)
```

```python
DATA_DIR = Path("../data").resolve()
```

```python
def node_degrees(G: Graph) -> dict[int, int]:
    return {i: G.degree(i) for i in G.nodes}
```

```python
def moment(data: Sequence[float], *, order: int = 1) -> float:
    return np.mean(np.pow(data, order))
```

```python
def graph_complexity(data: Sequence[float]) -> float:
    return moment(data, order=2) / moment(data, order=1)
```

```python
def calculate_distribution(data: Sequence[float]) -> dict[float, float]:
    counts = Counter(data)
    n = len(data)
    return {k: nk / n for k, nk in counts.items()}
```

```python
def shannon_entopy(data: Sequence[float]) -> float:
    distribution = calculate_distribution(data)
    probs = np.array([p for p in distribution.values()])
    return -probs.dot(np.log2(probs))
```

```python
def normalize_graph(G: Graph) -> Graph:
    G = G.to_undirected()
    G.remove_edges_from(nx.selfloop_edges(G))
    Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
    G = G.subgraph(Gcc[0])
    G = nx.convert_node_labels_to_integers(G, first_label=0)
    return G
```

## Questão 1

```python
G: Graph = nx.read_edgelist(DATA_DIR / "jazz.txt", nodetype=int)
G = normalize_graph(G)
```

```python
communities = greedy_modularity_communities(G)
```

```python
print(f"{modularity(G, communities) = :.2f}")
```

```python
mus = [0.05, 0.1, 0.2]
N = 128
tau1 = 3
tau2 = 1.5
k = 16
minc = 32
maxc = 32
seed = 10
```

```python
for mu in mus:
    G = nx.LFR_benchmark_graph(
        n=N,
        tau1=tau1,
        tau2=tau2,
        mu=mu,
        min_degree=k,
        max_degree=k,
        min_community=minc,
        max_community=maxc,
    )
    communities = louvain_communities(G)
    q = modularity(G, communities)
    print(f"Q({mu = :.2f}) = {q:.2f}")
```
